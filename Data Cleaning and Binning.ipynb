```python
data= dp['Salary in USD'].tolist()
def equal_frequency_binning(data, num_bins):
    data.sort()
    bins = []
    bin_size = len(data) // num_bins
    for i in range(0, len(data), bin_size):
        bins.append(data[i:i + bin_size])
    return bins
def smoothing_by_bin_means(bins):
    smoothed_bins = []
    for b in bins:
        mean = sum(b) / len(b)
        smoothed_bins.append([mean] * len(b))
    return smoothed_bins
def smoothing_by_bin_boundaries(bins):
    smoothed_bins = []
    for b in bins:
        min_val, max_val = b[0], b[-1]
        smoothed_bin = []
        for x in b:
            if x - min_val <= max_val - x:
                smoothed_bin.append(min_val)
            else:
                smoothed_bin.append(max_val)
                smoothed_bins.append(smoothed_bin)
                return smoothed_bins
num_bins = 3
bins = equal_frequency_binning(data, num_bins)
print("Equal-frequency bins:", bins)
smoothed_means = smoothing_by_bin_means(bins)
print("Smoothing by bin means:", smoothed_means)
smoothed_boundaries = smoothing_by_bin_boundaries(bins)
print("Smoothing by boundaries:", smoothed_boundaries)
```
